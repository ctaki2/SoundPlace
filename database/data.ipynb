```python
import pandas as pd
import numpy as np
import json

```

```python
data = pd.read_csv('pins.csv', on_bad_lines='skip', quotechar='"')
display(data)
print(data.shape)

```

```python
newPins = []

for index, row in data.iterrows():
    entry = {
        'id': index + 1,
        'license': row['license'],
        'artist': row['artistName'],
        'song': row['trackTitle'],
        'location': row['neighborhood'],
        'lat': row['latitude'],
        'lng': row['longitude'],
        'story': row['story'],
        'audio_url': row['trackStreamUrl'],
        'genre': row[data.columns[-1]]
        }
    newPins.append(entry)
```

```python
with open('pins.json', 'w') as f:
    json.dump(newPins, f, indent=4)
    
```
